### Meanings hunting






As we are trying to make the "universal" meaning extractor, we are free from any potential restrictions. Meaning is the more valuable entity for our task. So, to progress on the learning path to perfection our algorithm have to be greedy for new meainigs. In our case, it means that it looks for new context for every possible word in the clients messages. The more contexts it collects the more information it gets from any word, the better it understands the message. Because of this we have to constantly fill our corpus with the new texts that contain yet unseen contexts. Probably it should be implemented  through the api to some search engine. Aforementioned so called "meaning hunter" is so important of our task that I desided to name this engine as "Thompson".
When algorithm return wrong answer to the clients request this is the sign that we need get more meanings for every words in the request, and there is only one effective way to do this - get more texts with new context for this word.

### On the algorithms performance

Sometimes one picture can say more than thouthands of words, especially if the picture was drawn by Andrew Ng.
So here is a reference to diagram "Nuts and bolts of Applying Deep Learning" from the  talks at the Deep Learning School on September 24/25, 2016.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.transforms import BlendedGenericTransform

x = np.linspace(0.01, 6, 100)

fig,ax = plt.subplots(1)

trad_y =  x**(1/4)
small_y =  2*x**(1/4)
medium_y = 3*x**(1/3)
large_y = 5*x**(1/2)

plt.ylabel("Performance")
plt.xlabel("Data volume")
plt.title("Performance to data volume rate\n for different complexities algorithms")


#plt.axes.xaxis.set_ticklabels([])
#plt.axes.yaxis.set_ticklabels([])


plt.plot(x, trad_y, 'r--',  x, small_y, 'g--', x, medium_y,'b--', x, large_y, 'm--')

ax.set_yticklabels([])
ax.set_xticklabels([])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

plt.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off') 
plt.tick_params(axis='y',which='both',left='off',right='off',labelbottom='off')


ax.text(0, 1.05, r'$P$', transform=BlendedGenericTransform(ax.transData, ax.transAxes), ha='center')
ax.text(1.03, 0, r'$D$', transform=BlendedGenericTransform(ax.transAxes, ax.transData), va='center')

x_annot = 6.2
plt.text(x_annot, trad_y[-1], 'A - Traditional algorithms')
plt.text(x_annot, small_y[-1], 'B - Small NN')
plt.text(x_annot, medium_y[-1], 'C - Medium NN')
plt.text(x_annot, large_y[-1], 'D - Deep NN')

plt.show()
```

As we see the data volume is the matter, whatever algorithm we are going to use. However, the more complicated algorithm we apply, the more efficiently we use every bit of information. 

\begin{equation}
 \frac{\Delta P} {\Delta D} (A) < \frac{\Delta P} {\Delta D} (B) < \frac{\Delta P} {\Delta D} (C) < \frac{\Delta P} {\Delta D} (D)
\end{equation}

Equation shows us the more complicated algorithm we use the more efficient is the data usageÑŽ. The same applies and to NLP algoritms.
